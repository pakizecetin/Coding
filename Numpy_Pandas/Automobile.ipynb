```python
# Concatenate two data frames using the following conditions

GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV',
                          'Audi'], 'Price': [23845, 171995, 135925, 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan',
                            'Mitsubishi '], 'Price': [29995, 23600, 61500, 58900]}
german = pd.DataFrame(GermanCars)
japanese = pd.DataFrame(japaneseCars)
german
```

```python
japanese
```

```python
pd.concat([german,japanese], ignore_index=True)
```

```python
#Merge two data frames using the following condition
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV',
                         'Audi'], 'Price': [23845, 17995, 135925, 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda',
                              'BMV', 'Audi'], 'horsepower': [141, 80, 182, 160]}
```

```python
price =pd.DataFrame(Car_Price)
horsepower = pd.DataFrame(car_Horsepower)
price
```

```python
horsepower
```

```python
price.merge(horsepower)
```
